1.What is the role of filters and feature maps in Convolutional Neural Network (CNN)? 
Ans->
In a Convolutional Neural Network (CNN), filters and feature maps play a crucial role in extracting and representing information from input data, especially images. A filter, also known as a kernel, is a small matrix of weights that slides over the input image during the convolution operation. The main role of filters is to detect important patterns such as edges, textures, corners, and simple shapes present in the image. Each filter is designed to capture a specific type of feature, and its values are not fixed but learned automatically during the training process.

As the filter moves across the image, it performs element-wise multiplication with the corresponding region of the input and sums the result to produce a single value. This process is repeated over the entire image, resulting in a new matrix called a feature map. Therefore, a feature map is the output obtained after applying a filter to the input image.

Feature maps highlight the presence and location of the detected features in the input. If a certain pattern is strongly present in a region, the corresponding value in the feature map will be high. Each filter produces a separate feature map, and multiple filters are used in a CNN to capture different types of features simultaneously.

In the initial layers of a CNN, filters typically detect low-level features such as edges and lines. As we move deeper into the network, filters become more complex and start detecting high-level features like shapes, objects, and meaningful patterns. Thus, feature maps in deeper layers represent more abstract and informative features.

In summary, filters act as feature detectors that learn to identify important patterns, while feature maps store the results of these detections and pass them to the next layer for further processing. Together, they enable the CNN to understand and interpret visual data effectively.


2.Explain the concepts of padding and stride in CNNs(Convolutional Neural Network). How do they affect the output dimensions of feature maps?
Ans->
In Convolutional Neural Networks (CNNs), **padding** and **stride** are two important parameters that control how the convolution operation is applied to an input image and how the output feature map is generated. They directly affect the spatial size (height and width) of the resulting feature maps.

**Padding** refers to adding extra pixels (usually zeros) around the border of the input image before applying the filter. The main purpose of padding is to preserve the spatial dimensions of the input after convolution. Without padding, the size of the output feature map becomes smaller because the filter cannot be applied at the edges. There are different types of padding, such as “valid” padding (no padding) and “same” padding (padding added to keep output size equal to input size).

Padding also helps in retaining important edge information, which might otherwise be lost during convolution. By adding padding, we allow the filter to cover the border regions of the image more effectively.

**Stride** refers to the number of pixels by which the filter moves across the input image. A stride of 1 means the filter moves one pixel at a time, while a stride of 2 means it skips one pixel and moves two steps at a time. Increasing the stride reduces the size of the output feature map because fewer positions are covered.

Thus, stride controls the **downsampling** of the input. Larger stride values result in smaller feature maps and reduced computational cost, but may also lead to loss of information.

The output size of a feature map depends on input size, filter size, padding, and stride. Increasing padding increases or preserves the output size, while increasing stride decreases it.

In summary, padding helps control the border and preserve spatial dimensions, while stride controls the movement of the filter and reduces the size of the output. Together, they determine the final dimensions of the feature maps in a CNN.


3.Define receptive field in the context of CNNs. Why is it important for deep architectures?
Ans->
In the context of Convolutional Neural Networks (CNNs), the **receptive field** refers to the region of the input image that influences the value of a particular neuron in a feature map. In simple terms, it is the portion of the original input that a neuron “sees” or responds to.

In the first convolutional layer, the receptive field is usually small and equal to the size of the filter (for example, 3×3 or 5×5). This means each neuron only captures local features such as edges or textures. However, as we move deeper into the network, the receptive field increases. This happens because each neuron in deeper layers depends on a larger area of the input, due to multiple layers of convolution, pooling, and non-linear operations.

Thus, neurons in deeper layers have a **larger receptive field**, allowing them to capture more complex and global features such as shapes, objects, and patterns. The effective receptive field grows layer by layer, even if small filters are used.

The receptive field is important because it determines how much context a neuron can consider while making decisions. A small receptive field focuses on fine details, while a large receptive field captures broader spatial information.

In deep architectures, having a sufficiently large receptive field is essential for understanding the overall structure of the image. It helps the network combine local features into meaningful global representations.

If the receptive field is too small, the network may fail to capture important contextual information. On the other hand, if it is too large too quickly, it may lose fine details.

Therefore, CNNs are carefully designed to gradually increase the receptive field through multiple layers, allowing both local and global features to be learned effectively.

In summary, the receptive field defines the input area influencing a neuron, and it is crucial for enabling deep CNNs to learn hierarchical and meaningful representations of data.


4.Discuss how filter size and stride influence the number of parameters in a CNN.
 Ans->
In a Convolutional Neural Network (CNN), the number of parameters is mainly determined by the size and number of filters, while stride influences computation indirectly. Understanding how filter size and stride affect parameters is important for designing efficient models.

**Filter size** has a direct impact on the number of parameters. A filter is defined by its height, width, and depth (which matches the number of input channels). The total number of parameters in one filter is calculated as:
(height × width × input channels) + 1 bias term.
Therefore, larger filters (e.g., 5×5) have significantly more parameters than smaller filters (e.g., 3×3). For example, a 5×5 filter has 25 weights per channel, while a 3×3 filter has only 9, making it more efficient.

Using smaller filters reduces the number of parameters and helps in lowering computational cost. That is why modern CNN architectures often prefer multiple 3×3 filters instead of a single large filter, as they can achieve the same receptive field with fewer parameters and more non-linearity.

The **number of filters** also affects parameters. More filters mean more feature maps and therefore more total parameters in the network.

On the other hand, **stride** does not directly change the number of parameters. The filter weights remain the same regardless of how the filter moves across the input. However, stride affects the size of the output feature map.

When stride increases, the output feature map becomes smaller. This reduces the number of neurons in the next layer, which can indirectly reduce the total number of parameters in subsequent layers, especially in fully connected layers.

Thus, stride mainly controls computational efficiency and spatial resolution rather than parameter count.

In summary, filter size directly determines the number of parameters in a CNN, while stride indirectly affects the overall model size by changing the dimensions of feature maps and influencing later layers.
  

5.Compare and contrast different CNN-based architectures like LeNet, AlexNet, and VGG in terms of depth, filter sizes, and performance.
Ans->
Convolutional Neural Network (CNN) architectures such as **LeNet**, **AlexNet**, and **VGG** represent important milestones in the evolution of deep learning models for image recognition. They differ significantly in terms of depth, filter sizes, and performance.

**LeNet**, developed in the 1990s, is one of the earliest CNN architectures. It is relatively shallow, consisting of about 5–7 layers. It uses larger filters like 5×5 and was mainly designed for simple tasks such as handwritten digit recognition. Due to its simplicity, it has fewer parameters and lower computational requirements, but its performance is limited on complex datasets.

**AlexNet**, introduced in 2012, marked a breakthrough in deep learning. It is deeper than LeNet, with around 8 layers, and uses a mix of filter sizes such as 11×11, 5×5, and 3×3. It also introduced important techniques like ReLU activation, dropout, and GPU training. AlexNet significantly improved performance on large-scale image datasets like ImageNet, but it has a relatively large number of parameters.

**VGG**, developed later, further increased the depth to 16 or 19 layers. It uses a very uniform architecture with small 3×3 filters throughout the network. This design helps in capturing complex patterns while keeping the structure simple and consistent. However, VGG has a very high number of parameters, making it computationally expensive.

In terms of depth, LeNet is shallow, AlexNet is moderately deep, and VGG is very deep. In terms of filter size, LeNet uses larger filters, AlexNet uses mixed sizes, while VGG uses only small filters. In terms of performance, VGG generally outperforms the earlier models due to its depth and better feature extraction capability.

In summary, these architectures show a clear progression: from shallow and simple (LeNet) to deeper and more powerful (AlexNet) to very deep and highly accurate (VGG), highlighting the importance of depth and design choices in CNN performance.


```python
#6.Using keras, build and train a simple CNN model on the MNIST dataset from scratch. Include code for module creation, compilation, training, and evaluation.
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# Preprocessing
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Reshape to add channel dimension
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)

# One-hot encode labels
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

# Build CNN model
model = keras.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D(2, 2),
    
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Compile model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train model
model.fit(x_train, y_train, epochs=5, batch_size=64, validation_split=0.1)

# Evaluate model
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print("Test Accuracy:", test_accuracy)
```

```python
#7.Load and preprocess the CIFAR-10 dataset using Keras, and create a CNN model to classify RGB images. Show your preprocessing and architecture.
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Load CIFAR-10 dataset (32x32 RGB images, 10 classes)
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

# Preprocessing
# Normalize pixel values (0–255 → 0–1)
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# Convert labels to one-hot encoding
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

# Build CNN model
model = keras.Sequential([
    # Block 1
    layers.Conv2D(32, (3, 3), padding='same', activation='relu', input_shape=(32, 32, 3)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Dropout(0.25),

    # Block 2
    layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Dropout(0.25),

    # Fully Connected Layers
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])

# Compile model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Train model
model.fit(x_train, y_train,
          epochs=10,
          batch_size=64,
          validation_split=0.1)

# Evaluate model
test_loss, test_accuracy = model.evaluate(x_test, y_test)
print("Test Accuracy:", test_accuracy)

# Model summary (architecture)
model.summary()
```

```python
#8. Using PyTorch, write a script to define and train a CNN on the MNIST dataset. Include model definition, data loaders, training loop, and accuracy evaluation.
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# Device configuration (GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ----------------------------
# Data Loaders & Preprocessing
# ----------------------------
transform = transforms.Compose([
    transforms.ToTensor(),                  # Convert to tensor
    transforms.Normalize((0.1307,), (0.3081,))  # Normalize MNIST
])

train_dataset = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
test_dataset = datasets.MNIST(root="./data", train=False, download=True, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=1000, shuffle=False)

# ----------------------------
# CNN Model Definition
# ----------------------------
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        
        self.conv_layers = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=3, stride=1),  # 28x28 -> 26x26
            nn.ReLU(),
            nn.MaxPool2d(2),                            # 26x26 -> 13x13
            
            nn.Conv2d(32, 64, kernel_size=3, stride=1), # 13x13 -> 11x11
            nn.ReLU(),
            nn.MaxPool2d(2)                             # 11x11 -> 5x5
        )
        
        self.fc_layers = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 5 * 5, 128),
            nn.ReLU(),
            nn.Linear(128, 10)
        )

    def forward(self, x):
        x = self.conv_layers(x)
        x = self.fc_layers(x)
        return x

model = CNN().to(device)

# ----------------------------
# Loss & Optimizer
# ----------------------------
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# ----------------------------
# Training Loop
# ----------------------------
epochs = 5

for epoch in range(epochs):
    model.train()
    total_loss = 0
    
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    print(f"Epoch [{epoch+1}/{epochs}], Loss: {total_loss/len(train_loader):.4f}")

# ----------------------------
# Evaluation (Accuracy)
# ----------------------------
model.eval()
correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

accuracy = 100 * correct / total
print(f"Test Accuracy: {accuracy:.2f}%")
```

```python
#9.Given a custom image dataset stored in a local directory, write code using Keras ImageDataGenerator to preprocess and train a CNN model.
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models

# ----------------------------
# 1. Set dataset paths
# ----------------------------
train_dir = "data/train"   # change to your path
val_dir   = "data/val"     # change to your path

# ----------------------------
# 2. Image Preprocessing
# ----------------------------
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.2,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical'   # use 'binary' if 2 classes
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical'
)

# ----------------------------
# 3. CNN Model
# ----------------------------
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D(2,2),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(train_generator.num_classes, activation='softmax')
])

# ----------------------------
# 4. Compile
# ----------------------------
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# ----------------------------
# 5. Train
# ----------------------------
model.fit(
    train_generator,
    epochs=10,
    validation_data=val_generator
)

# ----------------------------
# 6. Evaluate
# ----------------------------
loss, acc = model.evaluate(val_generator)
print("Validation Accuracy:", acc)
```
